# Project 17: Calculate the Average of a List

```python
# This project teaches how to calculate the mean of numbers in a list
```

```python
# Step 1: Ask the user to enter numbers separated by commas
numbers_input = input("Enter numbers separated by commas: ")
```

```python
# Step 2: Convert the string into a list of numbers
numbers_list = [float(x) for x in numbers_input.split(",")]
```

```python
# Step 3: Calculate the sum of all numbers
total = 0
for num in numbers_list:
    total += num
```

```python
# Step 4: Calculate the average
average = total / len(numbers_list)
```

```python
# Step 5: Print the result
print(f"The average of the numbers is: {average}")
```
